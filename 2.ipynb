```python
import numpy as np
import pandas as pd
import sympy as sp

from utilities import *
```

```python
SLOPE_ANGLE = 3 * sp.pi / 360
DETECTOR_ANGLE = 2 * sp.pi / 3
CENTER_WATER_DEPTH = 120

PRECISION = 20

NM_TO_M = 1852
```

```python
# 创建一个 dataframe 用于存储数据
index = np.arange(0, 8)
angles = index * sp.pi / 4
columns = ['direction_angle', *np.arange(0.0, 2.4, 0.3)]
df = pd.DataFrame(index=index, columns=columns)
df['direction_angle'] = angles
df
```

```python
field = Field(1000, 1000, CENTER_WATER_DEPTH, SLOPE_ANGLE, PRECISION)

for i in index:
    # 读取方向角
    direction_angle = df['direction_angle'][i]

    for j in range(1, len(df.columns)):
        # 读取距离
        distance_from_center = df.columns[j] * NM_TO_M
        # 计算探测器坐标
        x = distance_from_center * sp.cos(direction_angle)
        y = distance_from_center * sp.sin(direction_angle)
        # 创建探测器
        detector = Detector(field, x, y, direction_angle=direction_angle, detector_angle=DETECTOR_ANGLE, precision=PRECISION)
        # 计算探测器扫描线段长度
        detector.get_scan_segment()
        df.iloc[i, j] = detector.scan_segment_length_proj_xy

df        
```

```python
df.to_csv('2.csv', index=False)
```
